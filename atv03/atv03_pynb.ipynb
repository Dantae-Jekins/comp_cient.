```python
import matplotlib.pyplot as plt
import numpy as np

mes = np.arange(12) + 1 # meses de 1 a 12
index = np.arange(12, dtype=np.uint32);

# quantidades de produtos vendidos por mês
creme_facial = np.array([ 2500, 2630, 2140, 3400, 3600, 2760, 2980, 3700, 3540, 1990, 2340, 2900 ])
limpeza_facial = np.array([ 1500, 1200, 1340, 1130, 1740, 1555, 1120, 1400, 1780, 1890, 2100, 1760 ])
pasta_dentaria = np.array([ 5200, 5100, 4550, 5870, 4560, 4890, 4780, 5860, 6100, 8300, 7300, 7400 ])
sabonete = np.array([ 9200, 6100, 9550, 8870, 7760, 7490, 8980, 9960, 8100, 10300, 13300, 14400 ])
shampoo = np.array([ 1200, 2100, 3550, 1870, 1560, 1890, 1780, 2860, 2100, 2300, 2400, 1800 ])
hidratante = np.array([ 1500, 1200, 1340, 1130, 1740, 1555, 1120, 1400, 1780, 1890, 2100, 1760 ])

# Gráfico 1 - Total de produtos Vendidos por mês - Linha
# Gráfico 2 - Gráfico com todos os produtos vendidos por mês - Linha
# Gráfico 3 - Comparativo de Creme Facial com Limpeza Facial por mês - Barras
# Gráfico 4 - Histograma de quantidade de meses (y) e faixas de quantidades de produtos vendidos (1000-1999, 2000-2999, ...)
# Gráfico 5 - Pizza. % da quantidade produtos vendidos no ano em cada produto
```

#### Gráfico 1 - Total de Produtos Vendidos por mês


```python
total_vendas = np.empty(12, dtype=np.uint32);
total_vendas[index] = (
      creme_facial[index] + 
      limpeza_facial[index] + 
      pasta_dentaria[index] + 
      sabonete[index] + 
      shampoo[index] + 
      hidratante[index]
      );

fig, graph = plt.subplots();
graph.plot(index, total_vendas, marker="s", label="Vendas");
graph.legend();
graph.set_title("Total de produtos vendidos por mês");
graph.set_xlabel("Mês");
graph.set_ylabel("Vendas");

fig.set_size_inches(10, 4);
fig.show();
```

#### Gráfico 2 - Gráfico com todos os produtos vendidos por mês

Hidratante e limpeza facial possuem os mesmos valores, por isso não é possível ver limpeza facial

```python
fig, graph = plt.subplots();
fig.set_size_inches(12, 6);
graph.plot(index, creme_facial, label="Creme facial");
graph.plot(index, limpeza_facial, label="Limpeza facial");
graph.plot(index, pasta_dentaria, label="Pasta dentária");
graph.plot(index, sabonete, label="Sabonete");
graph.plot(index, shampoo, label="Shampoo");
graph.plot(index, hidratante, label="Hidratante");
graph.set_xlabel("Mês");
graph.set_ylabel("Vendas");
graph.set_title("Comparação das vendas");
graph.legend()
fig.show()
```

#### Gráfico 3 - Comparativo de Creme Facial com Limpeza Facial por mês

```python
fig, graph = plt.subplots();
fig.set_size_inches(12, 6);

width = 0.4;
graph.bar(index - width/2, creme_facial, width = width, label="Creme facial");
graph.bar(index + width/2, limpeza_facial, width = width, label="Limpeza facial");
graph.set_ylabel("Vendas");
graph.set_title("Comparação das vendas entre Cremes e Limpeza facial");
graph.legend();
fig.show();
```

#### Gráfico 4 - Histograma de quantidade de meses (y) e faixas de quantidades de produtos vendidos

Não dava para fazer isso com os intervalos 0-1999 e tals, talvez se o considerado fosse a média aritmética e não a soma dos produtos vendidos, fiquei meio confuso xd.

```python
total_vendas = np.empty(12, dtype=np.uint32);
total_vendas[index] = (
      creme_facial[index] + 
      limpeza_facial[index] + 
      pasta_dentaria[index] + 
      sabonete[index] + 
      shampoo[index] + 
      hidratante[index]
      );

fig, histo = plt.subplots()
histo.hist(total_vendas);
histo.set_xlabel("Quantidade de produtos vendidos");
histo.set_ylabel("Quantidade de meses que venderam");
histo.set_title("Quantidade de meses que tiveram produtos vendidos nas faixas");
fig.show();
```

#### Gráfico 5 - Pizza, quantidade produtos vendidos no ano em cada produto

```python
total_dos_produtos = np.array([
      creme_facial.sum(), 
      limpeza_facial.sum(), 
      pasta_dentaria.sum(), 
      sabonete.sum(),
      shampoo.sum(), 
      hidratante.sum()]);
nome_dos_produtos = np.array([
    "Creme facial",
    "Limpeza facial",
    "Pasta dentária",
    "Sabonete",
    "Shampoo",
    "Hidratante",    
])

fig, histo = plt.subplots();
histo.pie(total_dos_produtos, labels=nome_dos_produtos, autopct="%1.1f%%");
histo.set_title("Porcentagem dos produtos vendidos em relação ao total");
fig.show();
```
